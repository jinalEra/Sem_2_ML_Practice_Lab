```python
bt = pd.DataFrame({
    "A": ['ani', 'ani', 'ani','bird', 'bird'],
    "B": ['one', 'one', 'two', 'one', 'two'],
    "C": ['small', 'large', 'large', 'large', 'small'],
    "D": [1,2,3,4,5],
    "E": [2,4,6,8,9]
})
bt
```

```python
import numpy as np
#pivot_table
table = pd.pivot_table(bt, values=['D','E'], index=['A','B'], columns=['C'], aggfunc= np.sum)
table
```

```python
t = pd.pivot_table(bt, values=['D'], index=['A'], columns=['C'],aggfunc=np.sum)
t
```
